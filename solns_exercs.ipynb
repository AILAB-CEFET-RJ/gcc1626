```python
import numpy as np

# --- Dados de entrada e saída desejada ---
x = np.array([[0.05], [0.10]])
y = np.array([[0.01]])

# --- Pesos e vieses iniciais ---
W1 = np.array([[0.15, 0.20],
               [0.25, 0.30]])
b1 = np.array([[0.35],
               [0.35]])
W2 = np.array([[0.40, 0.45]])
b2 = np.array([[0.60]])

# --- Funções de ativação ---
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_prime(z):
    s = sigmoid(z)
    return s * (1 - s)

# --- Forward pass ---
z1 = W1 @ x + b1
a1 = sigmoid(z1)
z2 = W2 @ a1 + b2
y_hat = sigmoid(z2)

# --- Cálculo da perda ---
L = 0.5 * (y_hat - y)**2

# --- Backward pass ---
delta2 = (y_hat - y) * sigmoid_prime(z2)
dW2 = delta2 @ a1.T
db2 = delta2

delta1 = (W2.T @ delta2) * sigmoid_prime(z1)
dW1 = delta1 @ x.T
db1 = delta1

# --- Atualização dos pesos ---
eta = 0.5
W2 -= eta * dW2
b2 -= eta * db2
W1 -= eta * dW1
b1 -= eta * db1

# --- Forward novamente após atualização ---
y_hat_new = sigmoid(W2 @ sigmoid(W1 @ x + b1) + b2)
L_new = 0.5 * (y_hat_new - y)**2

print("Perda antes da atualização:", L.item())
print("Perda depois da atualização:", L_new.item())
```
